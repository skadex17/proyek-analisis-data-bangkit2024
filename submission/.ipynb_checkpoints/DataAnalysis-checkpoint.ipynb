### Cleaning Data

Pada bagian ini kita akan melakukan pembersihan data dari permasalahan-permasalahan data yang sudah ditemukan pada tahap sebelumnya.

Pertama Kita akan mengubah nama-nama kolom yang ada di data day dan hour. Hal ini dilakukan untuk mempermudah membaca data yang ada di dalamnya untuk digunakan explore nanti. Hal ini dapat dilakukan dengan menggunakan rename seperti berikut :

```python
day_df = day_df.rename(
    columns={
        "dteday": "date",
        "yr": "year",
        "mnth": "month",
        "weekday": "day_of_week",
        "workingday": "workday",
        "weathersit": "weather",
        "temp": "temperature",
        "atemp": "apparent_temperature",
        "hum": "humidity",
        "casual": "casual_user",
        "registered": "registered_user",
        "cnt": "total_rental"
    }
)

hour_df = hour_df.rename(
    columns={
        "dteday": "date",
        "yr": "year",
        "mnth": "month",
        "hr": "hour",
        "weekday": "day_of_week",
        "workingday": "workday",
        "weathersit": "weather",
        "temp": "temperature",
        "atemp": "apparent_temperature",
        "hum": "humidity",
        "casual": "casual_user",
        "registered": "registered_user",
        "cnt": "total_rental"
    }
)
```

Setelah namanya diubah maka hasilnya akan seperti dibawah ini.

```python
day_df.head()
```

```python
hour_df.head()
```

Selanjutnya kita akan mengatasi masalah tipe data pada bagian data day dan hour. Untuk melakukan hal itu kita akan mengubah nilai integer pada data dan mengubahnya sesuai dengan nilai aslinya. Hal ini juga akan mempermudah explore data pada tahap selanjutnya. Untuk mengubahnya kita dapat menggunakan map atau replace, namun untuk sekarang kita akan menggunakan map seperti dibawah ini.

```python
season_mapping = {1: "Spring", 2: "Summer", 3: "Fall", 4: "Winter"}
year_mapping = {0: 2011, 1: 2012}
month_mapping = {1: "Jan", 2: "Feb", 3: "Mar", 4: "Apr", 5: "May", 6: "Jun", 7: "Jul", 8: "Aug", 9: "Sep", 10: "Oct", 11: "Nov", 12: "Dec"}
day_mapping = {0: "Sun", 1: "Mon", 2: "Tue", 3: "Wed", 4: "Thu", 5: "Fri", 6: "Sat"}
workday_mapping = {0: "No", 1: "Yes"}
holiday_mapping = {0: "No", 1: "Yes"}
weather_mapping = {1: "Clear", 2: "Cloudy", 3: "Light_rain", 4: "Heavy_rain"}

day_df["date"] = pd.to_datetime(day_df["date"])
day_df["season"] = day_df["season"].map(season_mapping).astype("category")
day_df["year"] = day_df["year"].map(year_mapping).astype("category")
day_df["month"] = day_df["month"].map(month_mapping).astype("category")
day_df["day_of_week"] = day_df["day_of_week"].map(day_mapping).astype("category")
day_df["workday"] = day_df["workday"].map(workday_mapping).astype("category")
day_df["holiday"] = day_df["holiday"].map(holiday_mapping).astype("category")
day_df["weather"] = day_df["weather"].map(weather_mapping).astype("category")

hour_df["date"] = pd.to_datetime(hour_df["date"])
hour_df["season"] = hour_df["season"].map(season_mapping).astype("category")
hour_df["year"] = hour_df["year"].map(year_mapping).astype("category")
hour_df["month"] = hour_df["month"].map(month_mapping).astype("category")
hour_df["day_of_week"] = hour_df["day_of_week"].map(day_mapping).astype("category")
hour_df["workday"] = hour_df["workday"].map(workday_mapping).astype("category")
hour_df["holiday"] = hour_df["holiday"].map(holiday_mapping).astype("category")
hour_df["weather"] = hour_df["weather"].map(weather_mapping).astype("category")
```

Dengan melakukan mapping diatas kita akan mengubah nilai integer pada masing-masing data sesuai dengan nilai yang kita assign. Kemudian kita menggunakan map tersebut pada semua kolom data yang ingin diubah. Dan pada bagian akhir kita akan mengubah tipe datanya menjadi category agar data dapat ditampilkan dan tidak menghasilkan missing value.

Setelah diubah hasilnya akan seperti dibawah ini.

```python
day_df.head()
```

```python
hour_df.head()
```

Selanjutnya sesuai dengan yang tertulis pada file readme, nilai pada kolom temperature, apparent_temperature, humidity, dan windspeed memiliki nilai maksimum yang berbeda-beda namun telah diubah menjadi rentang 0-1. Temperature memiliki nilai maksimal 41, apparent_temperature memiliki nilai maksimal 50, humidity memiliki nilai maksimal 100, dan windspeed memiliki nilai maksimal 67. Untuk mengubahnya kita hanya perlu mengalikan dengan nilai maksimal aslinya, seperti berikut :

```python
day_df["temperature"]=day_df["temperature"] * 41
day_df["apparent_temperature"]=day_df["apparent_temperature"] * 50
day_df["humidity"]=day_df["humidity"] * 100
day_df["windspeed"]=day_df["windspeed"] * 67

hour_df["temperature"] = hour_df["temperature"] * 41
hour_df["apparent_temperature"] = hour_df["apparent_temperature"] * 50
hour_df["humidity"] = hour_df["humidity"] * 100
hour_df["windspeed"] = hour_df["windspeed"] * 67
```

Untuk menampilkannya dapat dilakukan seperti di bawah ini.

```python
day_df.describe()
```

```python
hour_df.describe()
```

```python
hour_df.describe(include="all")
```

## Exploratory Data Analysis (EDA)

Pada bagian ini kita akan melakukan explore data yang sudah tersedia untuk mencari hubungan satu sama lain. Hal ini dapat membantu dalam menjawab pertanyaan bisnis yang ada pada bagian awal.

Pertama kita akan melihat hasil total rental pada setiap bulan.

```python
day_df.groupby(by="month", sort=False).agg({"total_rental": ["sum"]})
```

Seperti yang terlihat penjualan paling rendah berada pada bulan Januari dan Februari. Kemudian penjualan ini naik Pada bulan Maret dan April, dan ini terus mengalami peningkatan hingga puncaknya yaitu di bulan Agustus. Kemudian penjualan akan menurun perlahan terutama pada bulan November dan Desember. Disini dapat terlihat pola yang terbentuk, karena sepertinya penjualan meningkat atau turun tergantung pada musim.

Kita akan melihat penjualan musiman.

```python
day_df.groupby(by="season", sort=False).agg({"total_rental": ["sum"]})
```

Seperti yang dapat dilihat penjualan terbesar pada musim gugur, kemudian musim panas, musim dingin, dan yang terakhir musim semi.

```python
day_df.groupby(by=["month", "season"], sort=False).agg({"total_rental": ["sum"]})
```

Dapat kita lihat dari relasi bulan dan musim diatas bahwa rentang musim semi adalah dari Desember-Maret, rentang musim panas adalah dari Maret-Juni, rentang musim gugur adalah dari Juni-September, dan musim dingin memiliki rentang dari September-Desember. Dan dapat dilihat bahwa penjualan terbesar dapat dilihat dari rentang musim panas hingga awal musim dingin, dan penjualan terendah berada pada musim semi.

Selanjutnya kita ingin memastikan apakah suhu dapat menjadi faktor peningkatan pada setiap musim tersebut

```python
hour_df.groupby(by="season").agg({"temperature": ["min", "max"]})
```

```python
day_df.groupby(by="month", sort=False).agg({"temperature": ["min","max"]})
```

Jika dilihat dari hasil diatas, ditunjukkan bahwa musim dan bulan yang sebelumnya memiliki banyak penjualan memiliki suhu minimal dan maksimal yang tinggi. Dan bulan dimana penjualan menurun memiliki suhu minimal dan maksimal yang rendah

Selanjutnya kita akan mengecek faktor-faktor lain seperti cuaca, hari, hari kerja, hari libur, dan jam.

```python
hour_df.groupby(by="weather", sort=False).agg({"total_rental": "sum",})
```

Dapat dilihat cuaca berpengaruh besar pada penjualan, mengingat orang tidak akan bersepeda jika sedang hujan.

```python
day_df.groupby(by="workday", sort="false").agg({"total_rental":"sum"})
```

Data diatas menunjukkan bahwa penjualan pada hari kerja melebihi pada bukan hari kerja. Namun, perlu diingat bahwa jumlah hari kerja lebih banyak daripada jumlah bukan hari kerja.

```python
day_df.groupby(by="holiday").agg({"total_rental":"sum"})
```

Data di atas menunjukkan penjualan di hari biasa lebih banyak dibanding hari libur. Namun sama halnya seperti hari kerja, jumlah hari biasa lebih banyak daripada hari libur.

```python
hour_df.groupby(by="day_of_week", sort=False).agg({
    "total_rental": "sum",
    "temperature": ["min", "max"]
})
```

Seperti yang terlihat diatas penjualan setiap hari terdistribusi secara merata jika dibandingkan dengan data hari kerja dan hari libur sebelumnya. Dengan penjualan tertinggi berada di hari sabtu, kamis, dan jumat. Ini diasumsikan sama dengan pola sebelumnya yaitu suhu minimal dan maksimal yang lebih tinggi.

```python
hour_df.groupby(by="hour").agg({"total_rental":"sum", "temperature":["min", "max"]})
```

Pada data penjualan setiap jam, tampaknya penjualan tertinggi ada di jam 08.00, 17.00, 18.00. Disini pola suhu sepertinya tidak terlalu berpangaruh pada penjualan tiap jam. Bisa diasumsikan penjualan tertinggi terdapat pada jam tersebut dikarenakan waktu tersebut merupakan waktu awal beraktivitas dan akhir beraktivitas.

Mengingat faktor suhu yang memiliki pengaruh besar pada penjualan kita akan memastikan jumlah penjualan pada suhu tertentu yaitu :
jika dibawah sama dengan 10°C maka suhu tersebut termasuk rendah, jika diatas 25°C maka suhu tersebut termasuk tinggi, dan jika diantaranya maka akan masuk golongan sedang.

```python
day_df["temperature_level"] = day_df.temperature.apply(lambda x: "Low" if x <= 10 else ("High" if x > 25 else "Medium"))
hour_df["temperature_level"] = hour_df.temperature.apply(lambda x: "Low" if x <= 10 else ("High" if x > 25 else "Medium"))
```

```python
day_df.groupby(by="temperature_level", sort=False).agg({
    "total_rental": "sum"
})
```

jika dilihat maka penjualan tertinggi ada di suhu sedang, kemudian diikuti suhu tinggi dan suhu rendah. Ini membuktikan bahwa orang lebih suka memakai sepeda pada suhu sedang atau tinggi.

Terakhir kita akan melihat performa penjualan bisnis dalam 2 tahun

```python
day_df.groupby(by="year").agg({"total_rental":"sum"})
```

```python
day_df.groupby(by=["month", "year"], sort=False).agg({"total_rental":"sum"})
```

Dan seperti yang terlihat terdapat peningkatan signifikan yang terjadi pada performa bisnis pada tahun 2012 dibandingkan pada tahun 2011.

## Visualization & Explanatory Analysis

### Pertanyaan 1:

**Dalam rentang waktu mana bisnis ini mengalami kenaikan atau penurunan penjualan?**

Kita akan menampilkan data tersebut dalam bentuk bar plot

```python
monthly_rental = day_df.groupby(by="month", sort=False).agg({"total_rental": "sum"})

plt.figure(figsize=(10, 6))
plt.bar(monthly_rental.index, monthly_rental["total_rental"] / 1000, color='skyblue')  # divide by 1000 to maintain consistency
plt.xlabel('Month')
plt.ylabel('Total Rental (in thousands)')
plt.title('Total Rental by Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Pertanyaan 2:

**Apa yang menjadi faktor besar sebagai penyebab kenaikan atau penurunan penjualan dalam bisnis ini?**

Semua data disini akan ditampilkan dalam bentuk bar plot

```python
seasonly_rental = day_df.groupby(by="season", sort=False).agg({"total_rental": "sum"})

plt.figure(figsize=(10, 6))
plt.bar(seasonly_rental.index, seasonly_rental["total_rental"] / 1000, color='skyblue')  # divide by 1000 to maintain consistency
plt.xlabel('Season')
plt.ylabel('Total Rental (in thousands)')
plt.title('Total Rental by Season')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
day_df["temperature_level"] = day_df["temperature"].apply(lambda x: "Low" if x <= 10 else ("High" if x > 25 else "Medium"))

temperature_level_rental = day_df.groupby(by="temperature_level", sort=False).agg({"total_rental": "sum"})

plt.figure(figsize=(10, 6))
plt.bar(temperature_level_rental.index, temperature_level_rental["total_rental"] / 1000, color='skyblue')  # divide by 1000 to maintain consistency
plt.xlabel('Temperature Level')
plt.ylabel('Total Rental (in thousands)')
plt.title('Total Rental by Temperature Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
weatherly_rental = hour_df.groupby(by="weather", sort=False).agg({"total_rental": "sum"})

plt.figure(figsize=(10, 6))
plt.bar(weatherly_rental.index, weatherly_rental["total_rental"] / 1000, color='skyblue')
plt.xlabel('Weather')
plt.ylabel('Total Rental (in thousands)')
plt.title('Total Rental by Weather')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Pertanyaan 3:

**Bagaimana evaluasi performa bisnis ini dalam rentang 2 tahun?**

Untuk bagian ini kita akan membandingkan dua jenis line plot yaitu dari tahun 2011 dan dari tahun 2012

```python
monthly_rental = day_df.groupby(by=["month", "year"], sort=False).agg({"total_rental": "sum"})

monthly_rental["total_rental"] /= 1000

plt.figure(figsize=(10, 6))
for year in monthly_rental.index.get_level_values('year').unique():
    data_year = monthly_rental.xs(year, level='year')
    plt.plot(data_year.index.get_level_values('month'), data_year["total_rental"], label=str(year))

plt.xlabel('Month')
plt.ylabel('Total Rental (in thousands)')
plt.title('Total Rental by Month and Year')
plt.legend(title='Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Conclusion

### Conclusion Pertanyaan 1
**Dalam rentang waktu mana bisnis ini mengalami kenaikan atau penurunan penjualan?**

Berdasarkan data, rentang waktu kenaikan bisnis bike sharing ini adalah pada bulan Maret hingga pada puncaknya adalah pada bulan Agustus. Sedangkan, penurunan penjualan akan mulai dialami pada bulan September hingga Februari. Hal ini dipengaruhi oleh faktor seperti musim dan suhu. Mengetahui hal tersebut pemilik bisnis dapat mengantisipasi peningkatan jumlah pelanggan yang akan diterima.

### Conclusion Pertanyaan 2
**Apa yang menjadi faktor besar sebagai penyebab kenaikan atau penurunan penjualan dalam bisnis ini?**

Faktor terbesar yang mempengaruhi kenaikan atau penurunan penjualan adalah yang pertama musim. Pada data ditunjukkan bahwa musim gugur dan musim panas merupakan dimana penjualan akan naik. Sedangkan, pada musim dingin dan semi terjadi penurunan pada penjualan. Hal ini dipengaruhi lagi oleh faktor lain yaitu suhu. Data menunjukkan bahwa jauh lebih banyak orang yang lebih memilih menggunakan sepeda pada saat suhu sedang atau suhu tinggi. Faktor lain yang cukup signifikan adalah cuaca. Sesuai dengan data, cuaca hujan yang buruk dapat menyebabkan penurunan penjualan. Bahkan cuaca yang mendung memiliki dampak penjualan yang cukup besar dibandingkan cuaca cerah. Ini merupakan faktor yang harus selalu dimonitor oleh pemilik bisnis.

### Conclusion Pertanyaan 3
**Bagaimana evaluasi performa bisnis ini dalam rentang 2 tahun?**

Berdasarkan data, performa bisnis dari tahun 2011 dan tahun 2012 memiliki peningkatan yang sangat signifikan. Bahkan bagian dimana penjualan pada tahun 2011 mengalami penurunan, pada tahun 2012 performa mereka tetap meningkat. Ini menunjukkan bahwa terjadi sebuah progres perkembangan yang terjadi di dalam bisnis tersebut.
